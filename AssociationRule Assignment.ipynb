```python
def txt_split(txt):
    return txt.split(',')
```

```python
txt_split("shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil")
```

```python
data = df['Products'].apply(txt_split)
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
rules = pd.DataFrame(encoded_df,columns=te.columns_)
```

```python
te.columns_
```

```python
rules
```

```python
rules.replace([False,True],[0,1],inplace=True)
```

```python
rules
```

```python
pd.set_option('display.max_columns', None)
```

```python
scores = apriori(rules,min_support=0.01, use_colnames=True)
```

```python
scores
```

```python
score_rules = association_rules(scores,min_threshold=0.2)
```

```python
score_rules
```

```python
score_rules.sort_values(by = 'confidence', ascending=False)
```

We can see that if ground beef and eggs is purchased than mineral water will be purchased which is having 50% chance and also lift ratio 2.12 indicating both of them occurs
Similary if milk and ground beef is purchased then mineral water is also purchased with 50% confidence and lift 2.11
